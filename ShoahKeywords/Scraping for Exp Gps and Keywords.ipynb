```python
def find_index(list, sum):
    cumsum = 0
    for index, num in enumerate(list):
        cumsum += num
        if cumsum >= sum:
            return index
    return -1
```

```python
def adjMatrix(mat, t):
    adj_matrix = np.zeros_like(mat, dtype=int)
    adj_matrix[mat < t] = 1
    new_diagonal_value = 0
    np.fill_diagonal(adj_matrix, new_diagonal_value)
    return(adj_matrix)
```

```python
adjMatrix(distance_matrix, 0.15)
```
